```python
#read dna sequence
dna_sequence = "ATGCGACTACGAAAAGGTGGTAGATATATATATATATATATTTTAAATATATATATATAAAATTTATTACACATTCGAGGGCC"

#calculate the length
dna_length = len(dna_sequence)

#print the output
print("sequence is " + dna_sequence + " and length is " + str(dna_length))
```

```python
# dna sequence count the number of c and g
no_of_c = dna_sequence.count("C")

# dna sequence count the number of c and g
no_of_g = dna_sequence.count("G")

print("no of c is " + str(no_of_c) + " and no of g is " + str(no_of_g))
```

```python
# compute the percentage of GC content
gc_content = (no_of_c + no_of_g) / dna_length * 100 # percentage of GC content in the sequence

print("GC content is " + str(gc_content))
```

```python
# make it visual by using matplotlib library
import matplotlib.pyplot as plt

# create a list of labels
labels = "GC content", "AT content"

# create a list of values
sizes = [gc_content, 100 - gc_content]

# create a list of colors

colors = ["yellowgreen", "lightskyblue"]

# create a list of explode values

explode = (0.1, 0) # explode 1st slice

# create a pie chart

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=140)

# set aspect ratio to be equal so that pie is drawn as a circle

plt.axis("equal")

# show the plot

plt.show()


```
